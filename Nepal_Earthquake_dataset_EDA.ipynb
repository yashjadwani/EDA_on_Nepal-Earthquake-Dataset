## Understanding the dataset

The various aspects of building location and construction data consists of the following data columns:

- <b> building_id:</b> This column represents the Unique id of every buidling generated randomly
- <b> geo_level_1_id:</b> Largest geographic region in which building exists
- <b> geo_level_2_id:</b> Specific geographic region in which building exists
- <b> geo_level_3_id:</b> Most specific geographic sub-region in which building exists
- <b> count_floors_pre_eq:</b> Number of floors exists before earthquake
- <b> age:</b> This column shows the age of the building in years
- <b> area_percentage:</b> It shows the normalized area of the building footprint
- <b> height_percentage:</b> It shows the normalized height of the building footprint
- <b> land_surface_condition:</b> This columns helps in understating the surface condition of the land where the building was built
- <b> foundation_type:</b> It shows the foundation type used for construction of building
- <b> roof_type:</b> This column states the type of roof used while building
- <b> ground_floor_type:</b> It indicates the type of the ground floor
- <b> other_floor_type:</b> It indicates the type of construction used in higher floors other than the ground floor, excluding roof
- <b> position:</b> This column represents the position of the buidling
- <b> plan_configuration:</b> This column states the plan configuration of the structure
- <b> has_superstructure_adobe_mud:</b> Indicates if adobe mud was used for construction of the structure. 1=Yes , 0 = No
- <b> has_superstructure_mud_mortar_stone:</b> Indicates if mud mortar stone was used for construction of the structure. 1=Yes , 0 = No
- <b> has_superstructure_stone_flag:</b> Indicates if stone flag was used for construction of the structure. 1=Yes , 0 = No
- <b> has_superstructure_cement_mortar_stone:</b> Indicates if cement mortar was used for construction of the structure. 1=Yes , 0 = No
- <b> has_superstructure_mud_mortar_brick:</b> Indicates if mud mortar brick was used for construction of the structure. 1=Yes , 0 = No
- <b> has_superstructure_cement_mortar_brick:</b> Indicates if cement mortar brick was used for construction of the structure. 1=Yes , 0 = No
- <b> has_superstructure_timber:</b> Indicates if timber was used for construction of the structure. 1=Yes , 0 = No
- <b> has_superstructure_bamboo:</b> Indicates if bamboo was used for construction of the structure. 1=Yes , 0 = No
- <b> has_superstructure_rc_non_engineered:</b> This column says whether the superstructure was rc non engineered
- <b> has_superstructure_rc_engineered:</b> This columns says whether the superstructure is rc engineered
- <b> has_superstructure_other:</b> Indicates if super structure was made of any other material
- <b> legal_ownership_status:</b> It show the legal status of the building
- <b> count_families:</b> Number of families living in the building
- <b> has_secondary_use:</b> Indicates if the building was used for secondary use or not. 1=Yes, 0=No
- <b> has_secondary_use_agriculture:</b> Indicates if building was used for agriculture purposes. 1=Yes, 0=No
- <b> has_secondary_use_hotel:</b> Indicates if building was used as a hotel. 1=Yes, 0=No
- <b> has_secondary_use_rental:</b> Indicates if building was used for rental purposes. 1=Yes, 0=No
- <b> has_secondary_use_institution:</b> Indicates if building was used as any institution. 1=Yes, 0=No
- <b> has_secondary_use_school:</b> Indicates if building was used as school. 1=Yes, 0=No
- <b> has_secondary_use_industry:</b> Indicates if building was used as a industry. 1=Yes, 0=No
- <b> has_secondary_use_health_post:</b> Indicates if building was used as a health post. 1=Yes, 0=No
- <b> has_secondary_use_gov_office:</b> Indicates if building was used as a government office. 1=Yes, 0=No
- <b> has_secondary_use_use_police:</b> Indicates if building was used as a police station. 1=Yes. 0=No
- <b> has_secondary_use_other:</b> Indicates if building was used for any other secondary purpose. 1=Yes, 0=No
- <b> damage_grade:</b> This column represents the damage levels on building that was hit by the earthquake as follows: 
        1 represents low damage
        2 represents a medium amount of damage
        3 represents almost complete destruction

```python
# Removing the outlier value i.e 995
data = data[data.age<995] 

# Applying one of the Data wrangling technique to reset the index and getting a proper data after dropping inital index
data = data.reset_index() 
data = data.drop('index',axis=1)
```

```python
plt.boxplot(data.age,vert=False)
plt.title("Age after removing the outlier value(i.e. 995) but before taking natural log")
plt.show()

# Min Max Normalization of age
data['minmax_normalize_age'] =  (data.age - data.age.min())/(data.age.max() -  data.age.min())

plt.boxplot(data.minmax_normalize_age,vert=False)
plt.title("Age after removing the outlier value(i.e. 995) and after Min Max Normalization")
plt.show()
```

- There was no change in the age boxplot after normalization. In the boxplot of Min Max Normalization, the x-axis value changes from 0 to 1.

```python
fig, axs = plt.subplots(1, 2, figsize=(18, 8), sharey = True)
fig.suptitle('Age vs Min Max Normalize Age')
axs[0].hist(data.age)
axs[0].set_title('Histogram of Age')

axs[1].hist(data.minmax_normalize_age)
axs[1].set_title('Histogram of Age after Min Max Normalization')

plt.show()
```

- It is visible from the above histogram that the age is positively skewed. Inorder to achieve normal distribution, logarithmic transformation can be used. However, the given dataset consists of over 26000 newly built buildings (i.e age = 0) and log(0) is undefined, therefore, the use of logarithmic transformation is not possible.

##### 1. What is the impact of superstructure materials used on the damage level and its correlation with non-engineered and engineered reinforced concrete?

```python
# Creating a new dataframe of superstructure material only

buildings_with_different_superstructure_material = data[['has_superstructure_adobe_mud','has_superstructure_mud_mortar_stone',
        'has_superstructure_cement_mortar_stone','has_superstructure_mud_mortar_brick',
        'has_superstructure_cement_mortar_brick', 'has_superstructure_timber',
        'has_superstructure_bamboo', 'has_superstructure_rc_non_engineered',
        'has_superstructure_rc_engineered', 'has_superstructure_other']]
```

```python
# Creating a dictionary to store the value count of each the superstructure material used 

dictionary_to_store_total_values_of_damage_based_on_material_used = {}
for columns in buildings_with_different_superstructure_material.columns.tolist():
    val = data[columns].value_counts()[1]
    dictionary_to_store_total_values_of_damage_based_on_material_used[columns] = val 
```

```python
# Creating the dataframe from the dictionary created above
total_buildings_damage_on_different_superstructure_material = pd.DataFrame.from_dict(
                                                        dictionary_to_store_total_values_of_damage_based_on_material_used,
                                                        orient='index',columns=['total_damage_on_building'])
```

```python
total_buildings_damage_on_different_superstructure_material.total_damage_on_building.sort_values()
```

- Displaying the different superstructure material used for number of buildings. 

```python
# To plot the bar chart for the above dataframe

plt.barh(list(dictionary_to_store_total_values_of_damage_based_on_material_used.keys()),
         list(dictionary_to_store_total_values_of_damage_based_on_material_used.values()))
plt.xlabel('Number of buildings damaged')
plt.ylabel('Different superstructure materials')
plt.title('Damage on buildings constructed with different materials')
plt.show()
```

- In buildings built with materials like Mud mortar stone and Timber has undergone most damage. Therefore, we will plot the countplot of the most damaged superstructure with damage grade to identify any trend or special case. 

```python
# Creating a dataframes of the superstructure built using the matreials mud mortar stone and timber

buildings_with_mud_mortar_stone = data[data['has_superstructure_mud_mortar_stone']==1]
buildings_with_timber = data[data['has_superstructure_timber']==1]
```

```python
fig, axes = plt.subplots(1,2, figsize=(18,5))
fig.suptitle('Highest damage causing material')
sns.countplot(x='has_superstructure_mud_mortar_stone',data=buildings_with_mud_mortar_stone,
              hue='damage_grade',ax=axes[0])
axes[0].set_title('Superstructure made of Mud Mortar Stone')

sns.countplot(x='has_superstructure_timber',data=buildings_with_timber,
              hue='damage_grade',ax=axes[1])
axes[1].set_title('Superstructure made of Timber')

plt.show()
```

```python
fig, axes = plt.subplots(1,2, figsize=(18,5))
fig.suptitle('Damage when buildings are made up of Non-engineered and Engineered reinforced concrete')
sns.countplot(x='has_superstructure_rc_non_engineered',data=data[data['has_superstructure_rc_non_engineered']==1],
              hue='damage_grade',ax=axes[0])
axes[0].set_title('Has Superstructure Non-engineered reinforced concrete')

sns.countplot(x='has_superstructure_rc_engineered',data=data[data['has_superstructure_rc_engineered']==1],
              hue='damage_grade',ax=axes[1])
axes[1].set_title('Has Superstructure Engineered reinforced concrete')

plt.show()
```
